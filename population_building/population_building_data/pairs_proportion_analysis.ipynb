```python
class RandomSequenceGenerator():
    def __init__(self, arr):
        self.ptr_ = 0
        self.arr_ = arr

    def get_next(self):
        res = None
        if self.ptr_ < self.arr_.shape[0]:
            res = self.arr_[self.ptr_]
            self.ptr_ += 1
        return res

    def reload_seq(self, shuffle=False):
        self.ptr_ = 0
        if shuffle:
            np.random.shuffle(self.arr_)
```
