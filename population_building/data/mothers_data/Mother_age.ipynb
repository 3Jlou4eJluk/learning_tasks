```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BRaO2015-2021.txt')
```

```python
df
```

```python
df21 = df[df.Year == 2021]
```

```python
df_bryansk = df21[df21.Reg==1115]
df_bryansk
```

```python
df_bryansk.columns[3:44]
```

```python
urban_counts = df_bryansk.iloc[1, 3:].values
```

```python
n_childs = 2
row1 = 1
row2 = 2
lb = 3 + 41 * n_childs
rb = 3 + 41 * (n_childs + 1)
data_urban = df_bryansk.iloc[row1, lb:rb].values
data_rural = df_bryansk.iloc[row2, lb:rb].values
plt.figure(figsize=(15, 10))
plt.style.use('ggplot')
plt.bar(range(15, 56), data_urban.astype('int64'), label='urban')
plt.bar(range(15, 56), data_rural.astype('int64'), label='rural')
plt.legend()

plt.xticks(range(15, 56), [str(i) for i in range(15, 56)], rotation=45)

plt.title('Распределение возраста матери(' + str(n_childs) + '-й ребёнок)')
plt.xlabel('Возраст матери')
```
